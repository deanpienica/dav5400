## Answer 5
Here in first statement the price is divided with the carat correspondingly. ppc contants a column with the price per carat. 
Then in the second statement the query is set to check when the value of price/carat is maximum show the data.

```python
ppc=data['price'] / data['carat']
print(data.query('price/carat=='+str(ppc.max())))  
```

## Answer 6
First a array is created which contains the number of fair good very good premium and ideal diamonds are present.

Then after the cuts array is ready plt.bar function is called from matplotlib which plots the bar plot and show the frequncy of the cuts.

This graph shows that the frequency of the diamond increases with the increase in quality of cuts.

```python
cuts=[]
cuts.append(data.query("cut=='Fair'")['cut'].count())
cuts.append(data.query("cut=='Good'")['cut'].count())
cuts.append(data.query("cut=='Very Good'")['cut'].count())
cuts.append(data.query("cut=='Premium'")['cut'].count())
cuts.append(data.query("cut=='Ideal'")['cut'].count())
plt.bar(["Fair","Good","Very Good","Premium","Ideal"],cuts)
```

## Answer 7
First the variable data is copied to df.

This graph shows that the premium diamonds  starts from a very high range of price.

```python
df=data
df.boxplot(column='price', by='cut',figsize=(15,10))
```
