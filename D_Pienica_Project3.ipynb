## Data 1

## Test grades for students in match class

 Libraries to import

```python
import pandas as pd
```

 Read data

```python
data = pd.read_csv("https://raw.githubusercontent.com/deanpienica/dav5400/master/data1.csv")
```

```python
data.head(10)
```

 Handle nans

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.head(10)
```

 convert data to long form

```python
def gather( df, key, value, cols ):
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pd.melt( df, id_vars, id_values, var_name, value_name )
```

```python
data_long = gather( data, 'Test', 'Marks', ['Test1','Test2','Test3','Test4'] )
```

```python
data_long.head(10)
```

 Sort data

```python
data_long = data_long.sort_values(['Name','Test'])
```

 Tidy data in long format

```python
data_long
```

### Analysis 1 :Which student has the highest average test grades

```python
grouped_data = data_long.groupby(['Name']).mean()
print(grouped_data.sort_values('Marks',ascending=False))
```

### Data 3

## Land Area , Amentities, College Grads and Jobs



 Read data


```python


data = pd.read_csv("https://raw.githubusercontent.com/deanpienica/dav5400/master/data3.csv")
data.head(10)
```


 Convert data to long form


```python

def gather( df, key, value, cols ):
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pd.melt( df, id_vars, id_values, var_name, value_name )


data_long = gather( data, 'YearCategory', '#Grads', ['1980Grads','1990Grads'] )

data_long

data_long = gather( data_long, 'YearCategory', '#Jobs', ['1980Jobs','1990Jobs'] )

data_long
```


 Sort data

```python

data_long = data_long.sort_values(['County','LandArea'])

### Tidy data in long format

data_long
```

```python
### Analysis 1 : Do LandArea and Amentities rating lead to more grads and job
```

```python



grouped_data = data_long.groupby(['LandArea']).sum()
print(grouped_data.sort_values('#Jobs',ascending=False))

grouped_data = data_long.groupby(['Amentities']).sum()
print(grouped_data.sort_values('#Jobs',ascending=False))

grouped_data = data_long.groupby(['LandArea']).sum()
print(grouped_data.sort_values('#Grads',ascending=False))

grouped_data = data_long.groupby(['Amentities']).sum()
print(grouped_data.sort_values('#Grads',ascending=False))



```

 From the 4 cells above:
 
 Most grads and jobs in area with large land size
 
 Most grads and jobs in area with high amentities
 
 Hence LandArea and amentities have positive effect on number of grads and jobs


### Analysis 2 : Which county has most job to grad ratio (more oppurtunities, least competition) 



```python
grouped_data = data_long.groupby(['County']).sum()
(grouped_data['#Jobs'])

grouped_data = data_long.groupby(['County']).sum()
print(grouped_data['#Grads'])

ratio = grouped_data['#Jobs']/ grouped_data['#Grads']
print(ratio)

ratio.sort_values(ascending=False)

```

 Durham has highest job to grad ratio and is least competitive and Millwall is most compeitive
